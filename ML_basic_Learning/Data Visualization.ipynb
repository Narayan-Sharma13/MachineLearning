### Heat Graph

```python
import numpy as np

a=np.random.randint(1,100,25).reshape(5,5)
a
```

```python
corr = d1.corr()
ax = sns.heatmap(corr, square=True,annot=True)

ax.set_xticklabels(ax.get_xticklabels()) 
```

#### Here color intensity depands upon the value of the cell i.e. if the value is `lowest then color in darkest`

```python
sns.boxplot(y="total_bill",x="day",data=d1)
```

```python
# To show the uni-variate vs Bi-varaite Analysis under one plot
sns.jointplot(x="total_bill",y="tip",data=d1)
```

```python
sns.jointplot(x="total_bill",y="tip",data=d1,kind="hex") #'hex' = hexagons with dark area with highest density
```

```python
### To plot scatterplots for joint relationships and histograms for univariate distributions:


g = sns.pairplot(data=d1, vars = ["total_bill","tip"])
```
