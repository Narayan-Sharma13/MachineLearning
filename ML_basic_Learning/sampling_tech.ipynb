```python
# Importing the dataset
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
X=np.array(data)
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
```

```python
data['Labels']=pd.DataFrame(kmeans.labels_)
```

```python
data
```

```python
# plot
# Getting the values and plotting it
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
```

```python
a=data[data['Labels']==0].sample(n=3,random_state=3,replace=None)
b=data[data['Labels']==1].sample(n=3,random_state=3,replace=None)
c=data[data['Labels']==2].sample(n=3,random_state=3,replace=None)
##plot
##Getting the values and plotting it
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
plt.scatter(a['V1'],a['V2'])
plt.scatter(b['V1'],b['V2'])
plt.scatter(c['V1'],c['V2'])
```
