```python
def update_weights(m, b, X, Y, learning_rate):
    m_deriv = 0
    b_deriv = 0
    N = len(X)
    for i in range(N):
        # Calculate partial derivatives
        # -2x(y - (mx + b))
        m_deriv += -2*X[i] * (Y[i] - (m*X[i] + b))

        # -2(y - (mx + b))
        b_deriv += -2*(Y[i] - (m*X[i] + b))

    # We subtract because the derivatives point in direction of steepest ascent
    m -= (m_deriv / float(N)) * learning_rate
    b -= (b_deriv / float(N)) * learning_rate

    return m, b
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([0.5,2,2.5,3,3.5,2.75,3.25,1.3,1.4,1.7,1.6])
```

```python
Y=np.array([54,65,69,71,75,77,78,62,63,61,59])
```

```python
i=0
m,b=0.01,50
while i<400: 
    m,b=update_weights(m,b,X,Y,0.001)
    
    print('Iteration:{} {},{}'.format(i,m,b))
    i=i+1
    
```
